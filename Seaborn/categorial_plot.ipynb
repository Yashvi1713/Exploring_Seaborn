# Categorial Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Count Plot 

```python
# Here we have a supermarket data and we are trying to analyse that data using seaborn
mart_data = pd .read_excel('supermarket_sales.xlsx')
mart_data.info() 
# mart_data.head(5)
plt.figure(figsize=(15,5))

sns.countplot(x='Product line', data=mart_data)
```

```python
# Analysing the frequency of product line based on the gender

plt.figure(figsize=(15,5))
sns.countplot(x='Product line', hue="Gender", data=mart_data, palette="vlag")
```

#### Bar plot

```python
# Analysing total spend on per product line 
plt.figure(figsize=(15,5))

mart_sorted_data = mart_data['Product line'].sort_values()
mart_sorted_data.unique()
sns.barplot(x='Product line', y='Total', data=mart_data, palette='rocket', order=mart_sorted_data.unique())
```

```python
# Analysing spend of male & Female on individual product line
plt.figure(figsize=(15,5))
sns.barplot(x='Product line', y="Total", hue="Gender", data=mart_data, palette="rocket")
```

#### Box Plot

```python
# Box plot is mainly used to analyze the numaerical data as it shows key statistics like minimum, first Quartile(Q1: 25% of data)
# median, fourth Quartile(Q4: 75% of data)

sns.boxplot(y='Total', data=mart_data, showmeans = True, meanprops = {"marker":"o","markerfacecolor":"white","markersize":"10","markeredgecolor":"black"})
```

#### Violin plot

```python
# Violin plot is used when we are dealing with large datasets and where we have to compare density of data among different category
data = sns.load_dataset('tips')
data.columns
data.head()

sns.violinplot(x='day', y='total_bill',data=data, hue='sex', split='True')
```

#### Strip Plot

```python
# Strip plot is used when you have small amount of dataset and when you want to observe specific data point present in the category
#mart_data.info()  
plt.figure(figsize=(15,5))
sns.boxplot(x='Product line', y='Rating', data=mart_data, color='lightgray')
sns.stripplot(x='Product line', y='Rating', data=mart_data, jitter=0.1)
plt.show()
```

#### Swarm Plot

```python
# Swarm plot are similar to strip plot the only difference is that in swarm plot each data points are spreaded and they don't overlap like in strip plot

plt.figure(figsize=(10,3))
sns.boxplot(x='Payment',y='Total',data=mart_data)
sns.swarmplot(x='Payment',y='Total',data=mart_data)
plt.show()
```

#### Catplot

```python
# Cat plot allows us to visualize the different types of plot at one place. They are mainly used when we want to compare two different variables of the data
#mart_data.info()
plt.figure(figsize=(10,3))
sns.catplot(data=mart_data,x='Payment',y='Total',kind='violin', col='Gender')
plt.show()
```
