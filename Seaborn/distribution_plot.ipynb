```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

#### Histogram Plot

```python
mart_data = pd.read_excel('supermarket_sales.xlsx')
sns.histplot(data=mart_data,x='Total',binwidth=40,kde=True) # kde is kernel desnity estimation a way to estimate the probability density function of a continous random variable 
plt.show()
```

```python
# Hist stack plot can show how much each contribute to a whole, so we can say that for the transaction ranging from 0-200 has observed equal amount of payment method for all 3
sns.histplot(data=mart_data,x='Total',hue='Payment',multiple='stack',bins=5)
plt.show()
```

#### KDE(Kernel Density Estimation) plot

```python
# KDE plots are used for showcasing data distribution as well like histogram.
# When analysing large dataset then KDE plot is suitable as it provides smooth curve and easy to analyse whereas in historam are useful for smaller dataset and where there is more amount of modes bcz KDE plot will smoothen those modes and provide inacurrate observation.
# KDE plots shows relative probability values for the each interval

#negative_value = mart_data[mart_data['Total'] < 100]['Total']
#print(negative_value)
#mart_data[mart_data['Total'] < 100] => This creates a temporary dataframe for the contains thr rows which satisfies the condition total<100 and this is one of the way of filtering or manupilating data without changing the original dataframe mart_data   
#mart_data.shape => To ensure the original data is not changed

sns.kdeplot(data=mart_data,x='Total',hue='Payment',multiple='stack')
plt.show()

# From this we can say that the payment range 0-200 has equal proportion for each paymnet method
```

#### Rug Plot

```python
# Rug plot is used when we wan to see each data point on x or y axis or both as per the choice and with histogram or any other plot 
#mostly used for smaller dataset
# Generating random exam scores
np.random.seed(0)
exam_scores = np.random.randint(0, 101, size=100)
#print(exam_scores)

# Creating a histogram with a rug plot
#plt.figure(figsize=(10, 6))
sns.histplot(exam_scores, bins=10,color='skyblue',alpha=0.6)
sns.rugplot(exam_scores, color='red')

plt.title('Exam Scores Distribution with Rug Plot')
plt.xlabel('Exam Scores')
plt.ylabel('Density')
plt.show()
```

#### FacetGrid Plot

```python
# This plot is used when we want to compare different categories among the same variables to understand the differences in them
iris_data = sns.load_dataset('iris')
iris_data.info()

g = sns.FacetGrid(iris_data,col='species')
#g.map(sns.scatterplot,'sepal_length','sepal_width')
g.map(sns.scatterplot, 'sepal_length', 'petal_length', color='blue')
```

#### ECDF Plot (Empirical Cumulative Distribution Function )

```python
# It is more likely to use instead of the histogram for visualizing the data because
# the ECDF plot visualizes each and every data point of the dataset directly, 
# which makes it easy for the user to interact with the plot.

exam_scores =  np.random.randint(50,100,size=50)
#print(exam_scores)

sns.ecdfplot(exam_scores)
plt.show()
```

#### Displot

```python
# Displot is also use to observe the distribution of the data. It can plot 3 different plots i.e. Histogram,KDE,ECDF using kind parameter in it

tips_data = sns.load_dataset('tips')
#tips_data.info()

sns.displot(data=tips_data, x='total_bill',kind='hist',bins=5)
```

#### Joint Plot

```python
# This plot combines bivariate and univariate variables, making it possible to analyze raltionship between two variables and their individual distributions in the same plot.

#iris_data.info()
sns.jointplot(data=iris_data,x='sepal_length',y='petal_length',kind='scatter',hue='species')

# Plotting KDE and rug plot on top of joint plot from this it can be observed that the area in which kde is concentrated has cluster of data and the area disburst has less cluster of data
# Rug plot showcase the distribution on individual data points on the plot
l=sns.jointplot(data=iris_data,x='sepal_length',y='petal_length',joint_kws=dict(color='red'))
l.plot_joint(sns.kdeplot)
l.plot_joint(sns.rugplot,height=0.1,color='red')

plt.show()
```

#### Pair Plot

```python
# Pair plot is use to see the relationship between the two variables as well as there individual distribution at the same time
df = sns.load_dataset('tips')
selected_vars = ['total_bill', 'tip']
sns.pairplot(df, vars=selected_vars)
plt.show()
```
